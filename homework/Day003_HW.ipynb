### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？
from xml.dom.minidom import parse, parseString 
import xml.etree.ElementTree as et 
import pandas as pd

loc = d['cwbopendata']['dataset']['locations']['location']
# solution 1 
print(len(d['cwbopendata']['dataset']['locations']['location']))

# solution 2 
num = 0
for location in loc:
    num += 1
print(num)
    
# solution 3
df_data = pd.DataFrame()
for location in loc:
    list_A = []
    weatherElement = location['weatherElement']
    for weaele in weatherElement:
        if weaele['elementName'] == 'T':
            tm = weaele['time']
            for time in tm:
                list_A = []
                list_A.append(str(location['locationName']))
                list_A.append(location['geocode'])
                list_A.append(location['lat'])
                list_A.append(location['lon'])
                list_A.append(time['dataTime'])
                list_A.append(time['elementValue']['value'])
                df_data = df_data.append(pd.DataFrame(list_A).T)
        else:
            continue
df_data.columns = ['location', 'geocode', 'lat', 'lon', 'data_time', 'temp']
df_data.reset_index(drop = True, inplace = True)
print(df_data)
print(len(df_data.location.unique()))

```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

# solution 1 
key = d['cwbopendata']['dataset']['locations']['location']
for ky in key:
    weatherElement = ky['weatherElement']
    for weaele in weatherElement:
        if weaele['elementName'] == 'T':
            print(ky['locationName'])
            print('時間: ' + weaele['time'][0]['dataTime'] + ', 攝氏溫度: ' + weaele['time'][0]['elementValue']['value'] + '度')

# solution 2
    
```
