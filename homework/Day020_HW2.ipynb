```python
import requests
url = '	http://dcard.tw/_api/forums'

headers = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.61 Safari/537.36'}

r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text)
```
