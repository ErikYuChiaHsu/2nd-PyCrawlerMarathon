### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
list11 = []
for d in soup.find(class_="part_list_2").find_all('h3'):
    datetime = d.find(class_="date").text
    title = d.find_all('a')[-1].text.replace(u'\u3000',u' ')
    tag = d.find(class_='tag').text
    arr = [datetime, tag, title]
    list11.append(arr)
    # print(d.find(class_="date").text, d.find_all('a')[-1].text, d.find(class_='tag').text )
data = pd.DataFrame(list11, columns = ['datetime', 'tag', 'title'])
print(data)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
datetime, tag, title = [], [], []
list12 = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    datetime.append(d.find(class_="date").text)
    title.append(d.find_all('a')[-1].text.replace(u'\u3000',u' '))
    tag.append(d.find(class_='tag').text)
    
list12['datetime'] = datetime
list12['title'] = title
list12['tag'] = tag
print(list12)
```

### ③ 哪一個類別的文章最多

```python
tag_type_most = data['tag'].mode().values[0]

print(tag_type_most + ': ' + str(len(data[data['tag'] == tag_type_most])))
```
